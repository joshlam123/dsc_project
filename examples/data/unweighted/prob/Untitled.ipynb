```python
import json
import networkx
```

```python
# with open('data.txt') as json_file:
#     data = json.load(json_file)
with open('rand100.json') as f:
    file = json.load(f)

file
```

```python
import networkx as nx
G = nx.Graph()

for k,v in file['vertices'].items():
    G.add_node(v['Name'], value=v['Value'])
    for m in file['edges'][k]:
        G.add_edge(k, m['VerticeId'])
        
# nx.draw(G, with_labels=True, font_weight='bold')
options={"node_color":"blue"}
# nx.draw_shell(G, with_labels=True, font_weight='bold', **options)
# options = {'node_size': 20,'width': 3}

# nx.draw_spectral(G, **options)
nx.draw(G, with_labels=True, font_weight='bold')
# plt.subplot(122)
```
